```python
dmf1.head(10)
```

```python
data = [
    plotly.graph_objs.Bar(x=dmf1.index[0:30], y=dmf1['total'][0:30], name='症状'),
]
layout = plotly.graph_objs.Layout(
    title='症状のグラフ',
    xaxis={'title':'症状（英語）'}, 
    yaxis={'title':'数', 'rangemode':'tozero'},
)
fig = plotly.graph_objs.Figure(data=data, layout=layout)
fig.show()
```
