```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import plotly
```

```python
d1=pd.read_csv('/content/drive/MyDrive/2021VAERSData.csv',encoding='ISO-8859-1')
d2=pd.read_csv('/content/drive/MyDrive/2021VAERSSYMPTOMS.csv')
d3=pd.read_csv('/content/drive/MyDrive/2021VAERSVAX.csv',encoding='ISO-8859-1')
```

```python
d51=d1[['VAERS_ID','DIED','SEX']]
d52=d2[['VAERS_ID','SYMPTOM1','SYMPTOM2','SYMPTOM3','SYMPTOM4','SYMPTOM5']]
d53=d3[['VAERS_ID','VAX_TYPE','VAX_MANU']]
dm1=pd.merge(d51,d53,on='VAERS_ID')
dm2=pd.merge(dm1,d52,on='VAERS_ID')
dm2
```

```python
dm21=dm2.query('VAX_TYPE==["COVID19"]')
dm21
```

```python
dm22=dm21.query('DIED==["Y"]')
dm22
```

```python
dm21m=dm22.query('VAX_MANU==["MODERNA"]')
dm21m
```

```python
dm21mm=dm21m.query('SEX==["M"]')
dm21mm
```

```python
dm22m1=pd.DataFrame(dm21mm.SYMPTOM1.value_counts())
dm22m2=pd.DataFrame(dm21mm.SYMPTOM2.value_counts())
dm22m3=pd.DataFrame(dm21mm.SYMPTOM3.value_counts())
dm22m4=pd.DataFrame(dm21mm.SYMPTOM4.value_counts())
dm22m5=pd.DataFrame(dm21mm.SYMPTOM5.value_counts())
dm22mf=dm22m1.join(dm22m2)
dm22mf=dm22mf.join(dm22m3)
dm22mf=dm22mf.join(dm22m4)
dm22mf=dm22mf.join(dm22m5)
dm22mf=dm22mf.fillna(0)
dm22mf['total']=dm22mf['SYMPTOM1']+dm22mf['SYMPTOM2']+dm22mf['SYMPTOM3']+dm22mf['SYMPTOM4']+dm22mf['SYMPTOM5']
dm22mf['ratio']=dm22mf['total']/len(dm21mm)
dm22mf
dm22mff=dm22mf.sort_values(by=['total'],ascending=False)
dm22mfff=pd.DataFrame(dm22mff['total'])
dm22mffff=pd.DataFrame(dm22mff['ratio'])
dmfd1=dm22mfff.join(dm22mffff)
dmfd1.head(10)
```

```python
data = [
    plotly.graph_objs.Bar(x=dmfd1.index[0:10], y=dmfd1['total'][0:10], name='症状'),
]
layout = plotly.graph_objs.Layout(
    title='症状のグラフ',
    xaxis={'title':'症状（英語）'}, 
    yaxis={'title':'数', 'rangemode':'tozero'},
)
fig = plotly.graph_objs.Figure(data=data, layout=layout)
fig.show()
```
